```python
#titanic = sns.load_dataset('df')
#sns.countplot(df['duration'].value_counts())

plt.figure(figsize=(18, 10))
sns.countplot(x = 'duration', data = df, order = df['duration'].value_counts().index, palette ="autumn")
tick = plt.xticks(rotation='vertical')
plt.show()

# Best palettes for this:- spring_r (yellow pink), summer(green yellow), viridis_r(yellow green violet), winter_r(green blue), RdPu_r (purple white) , Wistia_r(Golden), autumn (red yellow),  
```
